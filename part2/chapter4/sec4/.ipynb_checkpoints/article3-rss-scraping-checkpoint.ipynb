feedparser는 굉장히 쉽게 사용할 수 있는 라이브러리입니다. parse 메서드에 url을 지정해서 호출하기만 하면 RSS피드를 읽어 들인 뒤 알아서 파싱해줍니다.

```python
import feedparser

rss = feedparser.parse('http://www.aladin.co.kr/rss/special_new/351')
print(rss)
```

```python
rss.get('feed')
```

```python
# 피드의 제목은 다음과 같이 확인 가능해요.
rss.get('feed').get('title')
```

```python
# 가장 중요한 RSS의 엔트리(item 요소)는 다음과 같이 확인합니다. 
rss.get('entries')
```

```python
# 신간 정보 엔트리의 타이틀과 링크를 출력해봅니다. 

for idx, content in enumerate(rss['entries']):
    print()
    print(str(idx+1)+'.',content.get('title'))
    print(content.get('link'))
    print()
```

```python
# rss version 확인
rss.version
```
